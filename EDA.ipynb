```python
# Step 1: Import essential libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# For displaying all columns without truncation
pd.set_option('display.max_columns', None)
```

```python
# For consistent plot style
sns.set(style="whitegrid")
```

```python
# Step 2: Load dataset

df = pd.read_csv("alzheimers_disease_data.csv")    
```

```python
# Preview top 5 rows
df.head()
```

```python
# Step 3: Basic dataset info

print("\nDataset Shape:", df.shape)
print("\nColumn Data Types:")
print(df.dtypes)

print("\nStatistical Summary:")
df.describe().T  
```

```python
# Step 4: Missing value check

missing = df.isnull().sum()
missing_percentage = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Values': missing,
    'Percentage (%)': missing_percentage
})

missing_df[missing_df['Missing Values'] > 0]
```

```python
# Step 5: Fill missing values
# Logic:
# - Numerical columns → median (robust against outliers)
# - Categorical columns → mode

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns

# Fill numeric columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Fill categorical columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode().iloc[0])
```

```python
df.isnull().sum().sum()
```

```python
# Age Distribution

plt.figure(figsize=(8,5))
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution of Patients")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# MMSE Score Distribution

plt.figure(figsize=(8,5))
sns.histplot(df["MMSE"], kde=True)
plt.title("MMSE Score Distribution (0–30)")
plt.show()
```

```python
# ADL (Daily Living Ability)
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["ADL"], kde=True)
plt.title("ADL Score Distribution (0–10)")
plt.show()
```

```python
# Diagnosis Count (Target Variable)

plt.figure(figsize=(6,4))
sns.countplot(x="Diagnosis", data=df)
plt.title("Diagnosis Distribution")
plt.xlabel("Diagnosis (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()
```

```python
# MMSE vs Diagnosis

plt.figure(figsize=(7,5))
sns.boxplot(x='Diagnosis', y='MMSE', data=df)
plt.title("MMSE vs Diagnosis")
plt.show()
```

```python
# ADL vs Diagnosis
plt.figure(figsize=(7,5))
sns.boxplot(x='Diagnosis', y='ADL', data=df)
plt.title("ADL Score vs Diagnosis")
plt.show()
```

```python
# Correlation Heatmap

plt.figure(figsize=(20, 12))

# Only use numeric columns for correlation

corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=False)
plt.title("Correlation Heatmap of Alzheimer’s Dataset")
plt.show()
```

```python
# Outlier Detection

cols_to_check = ['Age', 'BMI', 'CholesterolTotal', 'MMSE', 'ADL']

plt.figure(figsize=(12,6))
sns.boxplot(data=df[cols_to_check])
plt.title("Outlier Detection for Key Variables")
plt.show()
```

```python
# Distribution of Symptoms (Binary Features)

symptom_cols = [
    'Confusion','Disorientation','PersonalityChanges',
    'DifficultyCompletingTasks','Forgetfulness','BehavioralProblems'
]

df[symptom_cols].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,5))
df[symptom_cols].sum().sort_values().plot(kind='barh')
plt.title("Frequency of Symptoms in Dataset")
plt.show()
```

```python
# Feature vs Feature Insights
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age', y='MMSE', hue='Diagnosis', data=df)
plt.title("Age vs MMSE (Colored by Diagnosis)")
plt.show()
```
